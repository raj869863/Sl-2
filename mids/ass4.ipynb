```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Data points
points = np.array([
    [0.1, 0.6],   # P1
    [0.15, 0.71], # P2
    [0.08, 0.9],  # P3
    [0.16, 0.85], # P4
    [0.2, 0.3],   # P5
    [0.25, 0.5],  # P6
    [0.24, 0.1],  # P7
    [0.3, 0.2]    # P8
])

# Initial centroids: P1 and P8
initial_centroids = np.array([
    [0.1, 0.6],   # Cluster 1 start (P1)
    [0.3, 0.2]    # Cluster 2 start (P8)
])

# Run KMeans
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1, random_state=42)
kmeans.fit(points)
labels = kmeans.labels_
final_centroids = kmeans.cluster_centers_

# Plotting
colors = ['red', 'blue']
for i in range(2):
    cluster_points = points[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                c=colors[i], label=f'Cluster {i+1}')

# Plot initial centroids (green diamonds)
plt.scatter(initial_centroids[:, 0], initial_centroids[:, 1],
            c='green', marker='D', s=150, label='Initial Centroids')

# Plot final centroids (black X markers)
plt.scatter(final_centroids[:, 0], final_centroids[:, 1],
            c='black', marker='X', s=200, label='Final Centroids')

# Annotate points
for idx, (x, y) in enumerate(points):
    plt.annotate(f'P{idx+1}', (x + 0.005, y + 0.005))

plt.title('K-Means Clustering with Initial and Final Centroids')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.legend()
# plt.grid(True)
plt.show()
```
