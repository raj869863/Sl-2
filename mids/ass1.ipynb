```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
""" sir "cabin " has categorical data so it need not use mean/meadian 
    we can not use mode here --->> bcz large no. of null values
    we have to use forward fill " ffill()" """
```

```python
df['Cabin'] = df['Cabin'].ffill().bfill()
df
```

```python
df.isnull().sum()
```

```python
df['Embarked']= df['Embarked'].ffill()
df.isnull().sum()
```
