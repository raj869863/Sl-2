```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# def leaky_relu(x, alpha=0.01):
#     return np.where(x > 0, x, alpha * x)

# def elu(x, alpha=1.0):
#     return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Input Range
x = np.linspace(-10, 10, 400)

# Compute Activation Outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
# y_leaky_relu = leaky_relu(x)
# y_elu = elu(x)

# Plot Activation Functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, y_tanh, color='green')
plt.title("Tanh")
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, y_relu, color='red')
plt.title("ReLU")
plt.grid(True)

# plt.subplot(2, 3, 4)
# plt.plot(x, y_leaky_relu, color='purple')
# plt.title("Leaky ReLU")
# plt.grid(True)

# plt.subplot(2, 3, 5)
# plt.plot(x, y_elu, color='orange')
# plt.title("ELU")
# plt.grid(True)

plt.tight_layout()
plt.show()
```
